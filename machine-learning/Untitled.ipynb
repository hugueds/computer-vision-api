```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import pytesseract as pt
```

```python
file ='images/test-samples/IMG_20200527_154654883.jpg'
img = cv.imread(file, 0)
# img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img, cmap='gray')
```

```python
img_r = cv.resize(img, (480, 640), cv.INTER_AREA)
plt.figure(figsize=(12,8))
plt.imshow(img_r, cmap='gray')
```

```python
pt.image_to_string(img_r)
```

```python
croped = img_r[328:355, 238:350]
pt.image_to_string(croped)
plt.imshow(croped)
```

```python
_, thresh = cv.threshold(croped, 0, 255,cv.THRESH_BINARY | cv.THRESH_OTSU)
plt.imshow(thresh)
```

```python
pt.image_to_string(thresh)
```
